```python
import requests
import pandas as pd
import re

from bs4 import BeautifulSoup
```

```python
link_data = pd.read_excel("pilkada kabkot.xlsx", sheet_name="pilkada_kabkot")
link_data['new_link'] = link_data['link'].apply(lambda x: re.search(r'https://[^\s]+', x).group())
link_data
```

```python
link_kabkot = link_data['new_link'].to_list()
link_kabkot
```

```python
link_not_found = []
link_found = []

for link in link_kabkot:
    cek = re.search(r'https://[^\s]+&action=edit&redlink=1', link)
    
    if cek:
        link_not_found.append(cek.group())
    else:
        link_found.append(link)
```

```python
url = 'https://id.wikipedia.org/wiki/Pemilihan_umum_Bupati_Padang_Lawas_2024'
r = requests.get(url)
r
```

```python
soup = BeautifulSoup(r.content)
raw_data = soup.find('div',attrs={'id':'mw-content-text'})
ul = raw_data.find('ul')

if raw_data.find('ul'):
    print("ada")
else:
    print("tidak")
```

```python
data_dict = {'bakal_calon':[],
             'provinsi':[],
             'link_kabkot':[]}

for url in link_found:
    r = requests.get(url)
    soup = BeautifulSoup(r.content)
    raw_data = soup.find('div',attrs={'id':'mw-content-text'})
    ul = raw_data.find('ul')
    provinsi = re.search(r'Pemilihan_umum_[^\s]+',url).group()

    if ul :
        for u in ul:
            data_dict['bakal_calon'].append(u)
            data_dict['provinsi'].append(provinsi)
            data_dict['link_kabkot'].append(url)
    else:
        data_dict['bakal_calon'].append(None)
        data_dict['provinsi'].append(provinsi)
        data_dict['link_kabkot'].append(url)
```

```python
data_pilkada = pd.DataFrame.from_dict(data_dict, orient="index").T
data_pilkada = data_pilkada[data_pilkada['bakal_calon'] != '\n']
data_pilkada['bakal_calon'] = data_pilkada['bakal_calon'].astype(str)
data_pilkada.head()
```

```python
data_pilkada = pd.read_excel("result_sracp.xlsx")
data_pilkada.head()
```

```python
data_pilkada['new_bakal_calon'] = data_pilkada['bakal_calon'].apply(lambda x: re.findall(r'>([^<]+)</a>', str(x)) if re.search(r'>([^<]+)</a>', str(x)) else None)
data_pilkada.head()
```

```python
data_pilkada['nama_calon'] = data_pilkada['new_bakal_calon'].apply(lambda x: x[0] if x else None)
data_pilkada['pekerjaan'] = data_pilkada['new_bakal_calon'].apply(lambda x: x[1:] if x else None)
data_pilkada.head()
```
